# Try basics

```python
from graphviz import Digraph
from graphviz import render
from graphviz import Source

from pathlib import Path
```

```python
# Create Digraph object
dot = Digraph()

# Add nodes 1 and 2
dot.node('1')
dot.node('2')

# Add edge between 1 and 2
dot.edges(['12'])

dot
```

```python
# Add nodes 1 and 2
dot.node('Node A')
dot.node('Node B')
```

```python
dot
```

```python
dot.edges([('Node A', 'Node B'), ('1', 'Node A')])
dot
```

```python
print(dot.source)
```
