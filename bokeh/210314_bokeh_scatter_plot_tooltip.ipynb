```python
from math import sin
from random import random

from bokeh.io import output_file, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool, LinearColorMapper
from bokeh.palettes import plasma
from bokeh.plotting import figure
from bokeh.transform import transform

output_notebook()
```

```python
list_x = list(range(100))
list_y = [random() + sin(i / 20) for i in range(100)]
desc = [str(i) for i in list_y]

source = ColumnDataSource(data=dict(x=list_x, y=list_y, desc=desc))
hover = HoverTool(tooltips=[
    ("index", "$index"),
    ("(x,y)", "(@x, @y)"),
    ('desc', '@desc'),
])
mapper = LinearColorMapper(palette=plasma(256), low=min(list_y), high=max(list_y))

p = figure(plot_width=400, plot_height=400, tools=[hover], title="Belgian test")
p.circle('x', 'y', size=10, source=source,
         fill_color=transform('y', mapper))

# output_file('test.html')
show(p)
```

```python
list_x = list(range(100))
list_y = [random() + sin(i / 20) for i in range(100)]
desc = [str(i) for i in list_y]

source = ColumnDataSource(data=dict(x=list_x, y=list_y, desc=desc))
hover = HoverTool(tooltips=[
    #("index", "$index"),
    ("(x,y)", "(@x, @y)"),
    ('desc', '@desc'),
])
mapper = LinearColorMapper(palette=plasma(256), low=min(list_y), high=max(list_y))

p = figure(plot_width=400, plot_height=400, tools=[hover], title="Belgian test")
p.circle('x', 'y', size=10, source=source,
         fill_color=transform('y', mapper))

# output_file('test.html')
show(p)
```

```python
data2 = dict(x=list_x, y=list_y)

color = []

for x in data2['x']:
    temp_color = 'red' if x % 2 == 0 else 'green'
    # temp_color = (255, 0, 0, 0.5) if x % 2 == 0 else (0, 255, 0, 0.5)
    color.append(temp_color)

data2['color'] = color

# data2
```

```python
source = ColumnDataSource(data=data2)
hover = HoverTool(tooltips=[
    #("index", "$index"),
    ("(x,y)", "(@x, @y)"),
    ('color', '@color'),
])
mapper = LinearColorMapper(palette=plasma(256), low=min(list_y), high=max(list_y))

p = figure(plot_width=400, plot_height=400, tools=[hover], title="Belgian test")
p.circle('x', 'y', size=10, source=source,
         fill_color='color', fill_alpha=0.5, line_color=None)

# output_file('test.html')
show(p)
```
