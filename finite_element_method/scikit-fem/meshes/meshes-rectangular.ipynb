# `.init_tensor`

## Documentation

From [Class: MeshTri](https://scikit-fem.readthedocs.io/en/latest/api.html#class-meshtri):

```
Class: MeshTri
skfem.mesh.MeshTri
alias of MeshTri1

classmethod init_tensor(x: ndarray, y: ndarray)
Initialize a tensor product mesh.

The mesh topology is as follows:

*---------------*
|'-.|'-.|`'---._|
|---+---+-------|
|\  |\  |'.     |
| \ | \ |  '-.  |
|  \|  \|     '.|
*---------------*
Parameters:
x – The nodal coordinates in dimension x.

y – The nodal coordinates in dimension y.
```

## Manually create x & y arrays

### Equal x & y spacing

````python
x = np.array([0,1,2])
y = np.array([0,1])

MeshTri.init_tensor(x=x, y=y)
````

````python
MeshTri.init_tensor(x=x, y=y).refined(1)
````

````python
MeshTri.init_tensor(x=x, y=y).refined(2)
````

````python
m = MeshTri.init_tensor(x=x, y=y).refined(1)
m.save("meshes-rectangular1.vtk")
````

### Non-equal x & y spacing

````python
x = np.array([0,2])
y = np.array([0,1])

MeshTri.init_tensor(x=x, y=y)
````

````python
MeshTri.init_tensor(x=x, y=y).refined(1)
````

## Use np.linspace() to create x & y

````python
print(np.linspace(0, 2, 5))
print(np.linspace(0, 1, 3))
````

````python
MeshTri.init_tensor(np.linspace(0, 2, 5), np.linspace(0, 1, 3))
````

## Rectangular space with uniform cells

````python
x_dims = [-25, 25]
y_dims = [0, 10]
n_cells_x, n_cells_y = 5, 1
````

### MeshTri

````python
MeshTri.init_tensor(np.linspace(*x_dims, n_cells_x+1), np.linspace(*y_dims, n_cells_y+1))
````

````python
MeshTri.init_tensor(np.linspace(*x_dims, n_cells_x+1), np.linspace(*y_dims, n_cells_y+1)).refined(2)
````

### MeshQuad1

````python
MeshQuad1.init_tensor(np.linspace(*x_dims, n_cells_x+1), np.linspace(*y_dims, n_cells_y+1))
````

````python
MeshQuad1.init_tensor(np.linspace(*x_dims, n_cells_x+1), np.linspace(*y_dims, n_cells_y+1)).refined(2)
````
