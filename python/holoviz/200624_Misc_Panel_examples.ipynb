# Tap stream - get mouse click coordinates

See Philipp Rudiger's answer at [Example of using holoviews TapStream with Panel](https://discourse.holoviz.org/t/example-of-using-holoviews-tapstream-with-panel/166).

```python
import panel as pn
import holoviews as hv
import numpy as np

pn.extension()

points = hv.Points([])
stream = hv.streams.Tap(source=points, x=np.nan, y=np.nan)

@pn.depends(stream.param.x, stream.param.y)
def location(x, y):
    return pn.pane.Str(f'Click at {x:.2f}, {y:.2f}', width=200)

pn.Row(points, location)
```

# Checkbox

```python
checkbox = pn.widgets.Checkbox(name='Checkbox', value=True)

@pn.depends(value=checkbox)
def checkbox_value(value):
    return f"Checkbox value: {value}"

pn.Column(checkbox, checkbox_value)
```
