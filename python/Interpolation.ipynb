```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# %matplotlib widget

from scipy.interpolate import interp1d
import scipy
print(f"scipy version: {scipy.__version__}")
```

# scipy.interpolate.interp1d

## Sparse x,y data set

```python
x_vals = np.linspace(0, np.pi, 7)
y_vals = (1 + np.cos(x_vals)) / 2.

fig, ax = plt.subplots()
ax.plot(x_vals, y_vals)
ax.scatter(x_vals, y_vals);
```

## Create linear and cubic interpolation functions

In both cases, enable returning specified values outside of x range. Must set both `fill_value` and `bounds_error` in argument list.

```python
f_linear = interp1d(x_vals, y_vals, fill_value=(1,0), bounds_error=False)
f_cubic = interp1d(x_vals, y_vals, kind='cubic', fill_value=(1,0), bounds_error=False)
```

```python
def f_actual(x):
    return (1 + np.cos(x)) / 2.

x_new = np.linspace(0, np.pi, 101)
y_new_actual = f_actual(x_new)

x_beyond_range = np.linspace(-1, 4, 151)

fig, ax = plt.subplots(figsize=(10,8))
# ax.plot(x_new, y_new)
ax.plot(x_new, f_actual(x_new), label='Actual')
ax.plot(x_beyond_range, f_linear(x_beyond_range), label='Linear')
ax.plot(x_beyond_range, f_cubic(x_beyond_range), label='Cubic')
ax.legend();
```
